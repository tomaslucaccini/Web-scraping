# EJERCICIO 1

### Definición de Link al cual referenciar

```python
import requests

URL = 'https://www.bolsamadrid.es/esp/aspx/Indices/Resumen.aspx'
page = requests.get(URL).text
```

```python
page
```

### Transformación a formato BeautifulSoup

```python
soup = BeautifulSoup(page, 'lxml')
```

```python
soup
```

### Identificación de tabla en soup mediante ID

```python
tabla_indices = soup.find('table', {'id': 'ctl00_Contenido_tblÍndices'})
```

### Extracción de títulos

```python
titulos = []

for i in tabla_indices.find_all('th'):
    titulo = i.text
    titulos.append(titulo)
                
titulos
```

### Creación de DataFrame para posterior rellenado y dimensión de la iteración

```python
dataframe = pd.DataFrame(columns = titulos)

dataframe
```

### Carga de información en DataFrame

```python


for fila in tabla_indices.find_all('tr')[1:]:
    data = fila.find_all('td')
    data_fila = [td.text.strip() for td in data]
    largo = len(dataframe)
    dataframe.loc[largo] = data_fila
```

```python
dataframe
```
